```python
#Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
dataset= load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Filter df for 'Nepal' jobs and copy the result to df_np to prevent modification of original data.
df_np = df[df['job_country'] == 'Nepal'].copy()
df_np
```

```python
# Extract the full month name (e.g., 'January') from the 'job_posted_date' column
# and create a new column 'job_posted_month' to store it.

df_np['job_posted_month'] = df_np['job_posted_date'].dt.strftime('%B')

```

```python
df_np
```

```python
# Create a pivot table counting job postings by month (rows) and job title (columns).
df_np_pivot = df_np.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
df_np_pivot

```

```python
# Convert the 'job_posted_month' index back into a regular column.
df_np_pivot.reset_index(inplace=True)

```

```python

# Sort the pivot table rows chronologically by month using a temporary numeric month column,
# then set the month name back as the index and remove the temporary column.
df_np_pivot['month_no'] = pd.to_datetime(df_np_pivot['job_posted_month'], format='%B').dt.month
df_np_pivot['month_no'] # Displaying the temporary column (optional)
df_np_pivot.sort_values('month_no', inplace=True)
df_np_pivot.set_index('job_posted_month', inplace=True)
df_np_pivot.drop(columns='month_no', inplace=True)
```

```python
df_np_pivot
```

```python
# Create line Charts for each job title in the pivot table.
df_np_pivot.plot(kind='line')
```

```python

# Get the top 3 job titles based on the count of postings in the original DataFrame.
top_3= df_np['job_title_short'].value_counts().head(3)
top_3=top_3.index.tolist()
```

```python
# Plot the trends of the top 3 job titles over time.
df_np_pivot[top_3].plot(kind = 'line')
plt.title('Top 3 Job Posting Trends Over Time in Nepal', fontsize=14)
plt.xlabel('Posted Month', fontsize=12)
plt.ylabel('Number of Jobs', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.legend(title='Country')
plt.tight_layout()

plt.show()
```
