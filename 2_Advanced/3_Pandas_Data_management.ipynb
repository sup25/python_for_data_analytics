```python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()

# Print the memory address (ID) of the original DataFrame object.
print("ID of df_original:", id(df_original))

df_altered = df_original # Assigning without .copy() creates a reference, not a new object.

# Print the memory address (ID) of the referenced DataFrame; it will be the same as df_original's ID.
print("ID of df_altered:", id(df_altered))
```

```python
# df_altered now refers to the same DataFrame as df_original.
df_altered = df_original
# Print the ID (memory address) - will match df_original's ID.
print("ID of df_altered:", id(df_altered))

# df_altered now refers to a *new* copy of df_original.
df_altered = df_original.copy()
# Print the ID of the new copy - will be different.
print("ID of df_altered copy:", id(df_altered))
```
